```python
import matplotlib.pyplot as plt
import numpy as np

# Example movie data and sentiment scores
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D']
scores = [0.8, -0.5, 0.9, -0.3]

# Define colors based on sentiment scores
colors = ['green' if score >= 0 else 'red' for score in scores]

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot horizontal bars
barh = ax.barh(movies, scores, color=colors)

# Set labels and title
plt.xlabel('Sentiment Score')
plt.title('Sentiment Analysis of Movies')

# Add sentiment score values to the bars
for idx, bar in enumerate(barh):
    score = scores[idx]
    if score >= 0:
        ax.text(score + 0.05, idx, str(round(score, 2)), va='center')
    else:
        ax.text(score - 0.15, idx, str(round(score, 2)), va='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example movie sentiment scores
movies = ['Movie A', 'Movie B', 'Movie C']
positive_scores = [0.8, 0.6, 0.7]
negative_scores = [-0.5, -0.3, -0.6]
neutral_scores = [0.0, 0.1, -0.1]

# Create a stacked bar plot
fig, ax = plt.subplots()
bar_width = 0.3
index = np.arange(len(movies))

bar1 = plt.barh(index, positive_scores, bar_width, label='Positive', color='green')
bar2 = plt.barh(index, negative_scores, bar_width, label='Negative', color='red', left=positive_scores)
bar3 = plt.barh(index, neutral_scores, bar_width, label='Neutral', color='blue', left=np.array(positive_scores) + np.array(negative_scores))

plt.xlabel('Sentiment Score')
plt.ylabel('Movies')
plt.title('Sentiment Analysis of Movies')
plt.yticks(index, movies)
plt.legend()

plt.show()
```
